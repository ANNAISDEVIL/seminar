```python
import numpy as np
import matplotlib.pyplot as plt

# Pillow library, to manipulate images
from PIL import Image

def normalize_to_range(x, endrange):
    """ 
    Function that normalizes an input array or matrix to a desired range 
    using a linear transformation.
    
    A system of two linear equations that map from one range (max-min) 
    to the desired range 
    
    y_min = a*x_min + b
    y_max = a*x_max + b
    
    Two equations with two unknowns. Analytical solutions for a and b are:
    
    a = (y_max - y_min) / (x_max - x_min)
    b = y_max - a*x_max
    
    """
    
    a = (endrange[1] - endrange[0]) / (x.max() - x.min())  
    b = endrange[1] - a*x.max()
    
    return a*x+b 
```

### Reading image

```python
# reading image with PIL module
original = Image.open('orig1p9.png')
anime = Image.open('anig1p9.png')
# converting it to a numpy array. numpy arrays are easier to manipulate 
im_original = np.array(original)
im_anime = np.array(anime)

plt.imshow(im_original, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(im_anime, cmap='gray', vmin=0, vmax=255)

# size of full image
print(im_original.shape)
print(im_anime.shape)

# print(im_original[0])
```

### Mean squared error (MSE) calculation 

$$ MSE = \frac{1}{N} \sum^N_{i=1} (x_i - y_i)^{2}$$

with $N$ the number of pixels and $x$ being the original image and $y$ the distorted image.

```python
# definition of MSE
def mse(im1, im2):
    # size contains the size of a numpy array (total number of elements). 
    # shape contains the number of rows and columns
    N = im_original.size 
    return (1.0/N)*np.sum((im1 - im2)**2)


# showing images
def show_2ims(im, im_b):
    plt.figure()
    plt.subplot(1,2,1)
    plt.imshow(im, cmap='gray', vmin=0, vmax=255)

    plt.subplot(1,2,2)
    plt.imshow(im_b, cmap='gray', vmin=0, vmax=255)
    
    print('MSE', mse(im, im_b))
    
show_2ims(im_original, im_anime)
```

```python
def show_2cuts(im1, im2):    
    m = int(im_original.shape[0]/2) # middle row

    rgb1 = im1[m, :]
    rgb2 = im2[m, :]
    cut1 = []
    cut2 = []

    for key in rgb1:
        i = np.mean(key)
        cut1.append(i)
    for key in rgb2:
        i = np.mean(key)
        cut2.append(i)
    mean1 = np.mean(cut1)
    mean2 = np.mean(cut2)
    plt.figure()
    plt.plot(cut1, c='k', linewidth=1)
    plt.plot(cut2, c='b', linewidth=1)
    plt.axhline(y=mean1,color="black")
    plt.axhline(y=mean2,color="blue")
# cut1 is the mean value of rgb from original and cut2 for anime
    difference = mean1 - mean2
    print("Difference",difference)
    return difference
plt.show()
plt.imshow(im_original)
```

```python
delta = show_2cuts(im_original, im_anime)
print(delta)
```

```python
im_animex = im_anime+delta
newdelta = show_2cuts(im_original, im_animex)
print(newdelta)
```

```python
im_animex = im_animex.astype('uint8')
plt.imshow(im_animex, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(im_anime, cmap='gray', vmin=0, vmax=255)
plt.show()
plt.imshow(im_original)
```

### Manipulation e. white Gaussian noise contamination

```python
# we use the function normalize_to_range() to map the image values to the full range 0-255.
# the function does it using a simple linear transformation
im_original[:,:,0] = normalize_to_range(im_original[:,:,0], endrange = (50, 205))
im_original[:,:,1] = normalize_to_range(im_original[:,:,1], endrange = (50, 205))
im_original[:,:,2] = normalize_to_range(im_original[:,:,2], endrange = (50, 205))

im_animex[:,:,0] = normalize_to_range(im_animex[:,:,0], endrange = (50, 205))
im_animex[:,:,1] = normalize_to_range(im_animex[:,:,1], endrange = (50, 205))
im_animex[:,:,2] = normalize_to_range(im_animex[:,:,2], endrange = (50, 205))
```

```python
print(im_original.min())
print(im_original.max())
print(im_animex.min())
print(im_animex.max())

plt.imshow(im_original)
#plt.imshow(im_anime)

Image.fromarray(im_original).save('orig3p9.png', 'png')
Image.fromarray(im_animex).save('anig3p9.png', 'png')
```

```python
# adds zero-mean, gaussian distributed noise with s.d. equals to sigma (the amount of noise)
sigma = 5
# noise = np.random.normal(loc=0, scale=sigma, size=im_original.shape)
noise = np.random.normal(loc=0, scale=sigma, size=im_original.shape)
# noise = noise.astype('uint8')

print(noise.shape)
plt.figure()
plt.imshow(noise, cmap='gray')

print(noise.min())
print(noise.max())
```

```python
im_e1 = im_original + noise 
im_e2 = im_animex + noise

im_e1 = im_e1.astype('uint8')
im_e2 = im_e2.astype('uint8')

#print(im_e1)

print(im_e1.min())
print(im_e1.max())

plt.figure()
show_2ims(im_original, im_e1)
plt.show()
show_2ims(im_animex, im_e2)
```
